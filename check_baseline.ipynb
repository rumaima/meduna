```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_theme()
import pickle
```

```python
from sklearn.metrics import f1_score, roc_auc_score
```

```python
dataset_name = "ShenzhenCXR"

# load pickle file in a dicitonary called data
Hayat_pickle = f"/l/users/umaima.rahman/datasets/dataloaders/Hayat_pickle/{dataset_name}_hayat.pkl"

# Load the pickle file
with open(Hayat_pickle, 'rb') as f:
    data = pickle.load(f)

acc_vec = []
f1_vec = []
auc_vec = []
xrange_ = np.arange(0.1, 0.9, 0.01)
for threshold in xrange_:
    y_pl = (np.max(data['y_hat'], axis=1)> threshold).astype('int16')
    acc = np.mean((y_pl == data['y_true']).astype('int16'))
    f1 = f1_score(data['y_true'], y_pl)
    auc = roc_auc_score(data['y_true'], y_pl)
    acc_vec.append(acc)
    f1_vec.append(f1)
    auc_vec.append(auc)

print(dataset_name)
print(xrange_[np.argmax(acc_vec)], np.max(acc_vec) , f1_vec[np.argmax(acc_vec)], auc_vec[np.argmax(acc_vec)] )
```
